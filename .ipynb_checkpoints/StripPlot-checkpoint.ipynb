```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.stripplot(x='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips)
```

```python
sns.stripplot(x='day',y='tip',data=tips,jitter=True)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,jitter=True)
```

```python
sns.stripplot(x='total_bill',y='day',data=tips,jitter=True,linewidth=1.2)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='sex',dodge=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='smoker',dodge=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='smoker',dodge=True,order=['Sat','Sun','Thur','Fri'])
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='smoker',dodge=True,marker='D',order=['Sat','Sun','Thur','Fri'])
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='smoker',dodge=True,marker='D',size=10,order=['Sat','Sun','Thur','Fri'])
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,linewidth=1.2,hue='smoker',dodge=True,marker='D',size=10,order=['Sat','Sun','Thur','Fri'],edgecolor='gray',alpha=0.2)
```

```python
sns.stripplot(x='tip',y='day',jitter=True,data=tips)
sns.violinplot(x='tip',y='day',color='0.9',data=tips)
```
